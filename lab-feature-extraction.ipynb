```python
# Plot time variable. Can you extract something from it?
# Not clear, so I like to extract only the months
plt.figure(figsize=[10,8])
n, bins, patches = plt.hist(data['Effective To Date'],50,  density=False, facecolor='y', alpha=1, stacked=True)
plt.xlabel('Effective To Date')
plt.ylabel('number of counts')
plt.title('Effective To Date')

plt.show()
```

```python
DateMonth=[]
#data['Effective To Date'].value_counts()
for i in data['Effective To Date']:
    j = pd.to_datetime(i, format ="%m/%d/%y")
    DateMonth.append(pd.to_datetime(j, format='%m/%d/%y').strftime('%B'))
data['effective_to_date_Month']=DateMonth

data['effective_to_date_Month'].value_counts()
```

```python
data['effective_to_date_Month'].value_counts().plot(kind='bar')
plt.xlabel('month')
plt.ylabel('number of counts')
plt.title('Effective_to_date_Month')
# there are more counts for january, but not much. It might be better to extract not that much as it dont
# give a lot of information. It might be better to check out the weekdays
```
